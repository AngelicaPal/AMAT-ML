## lectura de archivo .xlsx

## archivo .pkl 

```python
## escritura
with open('data/ames.pkl', 'wb') as f:
    pickle.dump(ames_xlsx, f, pickle.HIGHEST_PROTOCOL)
```

```python
## lectura
with open('data/ames.pkl', 'rb') as f:
    ames_pkl = pickle.load(f)

ames_pkl.head(3)
```
