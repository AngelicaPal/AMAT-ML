```python
import pandas as pd
from siuba import *
```

```python
## Leemos el insumo
ames_housing = pd.read_csv("data/ames/ames.csv")
```

```python
ames_housing.info()
```

```python
ames_housing.describe()
```

## Selección de columnas 


```python
(
ames_housing >>
  select(_.Lot_Area, _.Neighborhood, _.Year_Sold, _.Sale_Price)
)
```

```python
## selecciona las columnas que contenga "Area", en cualquier lugar
(
 ames_housing >> 
  select(_.contains("Area"))
 )
```

```python
## selecciona las columnas que termine con "Area"
(
 ames_housing >> 
 select(_.contains("Area$"))
)
```

```python
## selecciona las columnas que comience con "Garage"
(
 ames_housing >> 
 select(_.contains("^Garage"))
)
```

## Filtrar observaciones

```python
ames_housing['Sale_Condition'].value_counts()
```

```python
(
  ames_housing >> 
  filter(_.Sale_Condition == "Normal")
)
```

```python
(
  ames_housing >> 
  filter( (_.Lot_Area > 1000) & (_.Sale_Price >= 150000) )
)
```

```python
(
  ames_housing >> 
  filter((_.Lot_Area < 1000) | (_.Sale_Price <= 150000))
)
```

```python
(
ames_housing >>
  filter( ((_.Street.isin(["Pave", "Grvl"]) ) & (_.Sale_Price < 10000000)) | (_.Lot_Area < 1000) )
)
```

#### Ordenar registros ####

```python
## ascendente
(
  ames_housing 
  >> arrange(_.Sale_Price)
  >> select(_.Sale_Price)

)
```

```python
## Descendente
(
  ames_housing >> 
  arrange(-_.Sale_Price)
  >> select(_.Sale_Price)
)
```

```python
(
ames_housing >> 
 arrange(_.Sale_Condition, -_.Sale_Price, _.Lot_Area) >>
 select(_.Sale_Condition, _.Sale_Price, _.Lot_Area)
)
```

#### Agregar o modificar columnas ####


```python
## Se crea una nueva columna "Antique"
ejemplo_mutate = ( 
 ames_housing >> 
   select(_.Year_Sold, _.Year_Remod_Add) >>
   mutate(Antique = _.Year_Sold - _.Year_Remod_Add)
)
## se debe almacenar en la variable "ejemplo_mutate"
```

```python
ejemplo_mutate
```

```python
## modificamos la columna Antique, multiplicamos la antiguedad para tenerla en meses
(
ejemplo_mutate >> 
 mutate(Antique = _.Antique * 12)
)
```

#### Agregaciones ####

```python
(
ames_housing >> 
 select(_.Year_Sold, _.Year_Remod_Add) >>
 mutate(Antique = _.Year_Sold - _.Year_Remod_Add) >>
 summarize(
  Mean_Antique = _.Antique.mean(),
  Median_Antique = _.Antique.median(),
  First_Antique = _.Antique.iloc[0],
  Last_Antique = _.Antique.iloc[-1],
  )
)
## summariza es una agregación quete permite calcular valores estadisticos
```

```python
(
ames_housing >> 
 select(_.Year_Sold, _.Year_Remod_Add) >>
 mutate(Antique = _.Year_Sold - _.Year_Remod_Add) >>
 arrange(_.Antique) >>
 summarize(
  Mean_Antique = _.Antique.mean(),
  Median_Antique = _.Antique.median(),
  min_Antique = _.Antique.iloc[0],
  max_Antique = _.Antique.iloc[-1],
  )
)
```

#### Agrupamiento ####

```python
(
ames_housing >> 
 mutate(Antique = _.Year_Sold - _.Year_Remod_Add) >> 
 group_by(_.Neighborhood) >> 
 summarize(Mean_Antique = _.Antique.mean().round(0) )
)
```

```python
(
ames_housing >> 
 mutate(Antique = _.Year_Sold - _.Year_Remod_Add) >> 
 group_by(_.House_Style) >> 
 summarize(Mean_Sale_Price = _.Sale_Price.mean().round(0)) >>
 arrange(_.Mean_Sale_Price)
)
```
