```python
from siuba import *
import plotnine as p9
from plotnine import aes, element_blank, data as p9d
from mizani.formatters import comma_format, dollar_format
from sklearn import datasets
```

```python
diamonds = p9d.diamonds
diamonds
```

```python
(
diamonds >>
  p9.ggplot(aes(x = "price")) +
  p9.geom_histogram(color = "pink", fill = "purple", bins=30) +
  p9.scale_x_continuous(labels=dollar_format()) + 
  p9.scale_y_continuous(labels=comma_format()) + 
  p9.ggtitle("DistribuciÃ³n de precio")
)
```

```python
(
 diamonds >>
  p9.ggplot(aes(x = "price")) +
  p9.geom_histogram(aes(y='stat(density)'), bins = 30, fill = 'blue', color = "white") +
  p9.geom_density(colour = "black", size = 1)
)
```

```python
(
diamonds >>
 p9.ggplot() +
 p9.geom_boxplot(aes(x = 0, y = "price"), color= "blue", fill= "lightblue") +
 p9.scale_y_continuous(labels = dollar_format(prefix='$', digits=0, big_mark=',')) +
 p9.theme(axis_text_x=element_blank()) +
 p9.ggtitle("DistribuciÃ³n de precio")
)
```

```python
(
diamonds >>
 p9.ggplot(aes(x = 0, y = "carat")) +
 p9.geom_boxplot(color= "purple", fill= "pink", alpha = 0.3) +
 p9.scale_y_continuous(labels = comma_format(digits=1)) +
 p9.theme(axis_text_y=element_blank()) +
 p9.coord_flip() +
 p9.ggtitle("DistribuciÃ³n del peso de los diamantes")
)
```
